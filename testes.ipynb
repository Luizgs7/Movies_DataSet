```python
from pandasql import sqldf
import pandas as pd

df_movies = pd.read_csv('Total/movies_df.csv', sep=';')
df_genres = pd.read_csv('Total/movies_genres.csv', sep=';')

query = """   
       SELECT  * FROM (
              SELECT DISTINCT
                     T2.imdb_id, T2.poster_path, T2.genres_name, row_number() over(order by ABS(RANDOM())) AS ORDEM,adult, vote_count
              FROM (SELECT 
                            T0.imdb_id,
                            T0.poster_path,
                            T1.genres_name,
                            vote_count,
                            adult,
                            row_number() over(partition by T0.imdb_id) AS RK
                     FROM df_movies AS T0
                     INNER JOIN df_genres AS T1
                     ON T0.imdb_id = T1.imdb_id 
                     WHERE T0.poster_path IS NOT NULL
                     AND CAST(SUBSTR(T0.release_date,7,4) AS INTEGER) BETWEEN 2010 AND 2020
                     AND T0.original_language = 'en'
                     AND T0.adult = 0
                     AND CAST(vote_count AS INTEGER) > 10
                     ) AS T2
              WHERE RK = 1
       )
       """

df = sqldf(query)
df
```
